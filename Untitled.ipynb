```python
# Scraping Code
import requests
from bs4 import BeautifulSoup

url = "https://medium.com/@pythoncodelab/building-a-to-do-list-app-in-python-a-step-by-step-guide-ce34b9ea141a"
page = requests.get(url)
soup = BeautifulSoup(page.text, "html")
print(soup)
```

```python
info = soup.find("span", id="e8b2")
print(info)
```

```python
for br in info.find_all("br"):
    br.replace_with("\n")
```

```python
final_output = info.get_text()

with open("to-do-2.py", "w+") as file:
    file.write(final_output.strip())
```
